```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('../datos/result/datos_brasil_unificado.csv')
```

```python
df.dtypes
```

```python
df['VALOR_PREVISTO_ACTUALIZADO'].unique()
```

```python
# Reemplazamos las comas por puntos.
df['VALOR_PREVISTO_ACTUALIZADO'] = df['VALOR_PREVISTO_ACTUALIZADO'].str.replace(',', '.')

# Convertir la columna al tipo float
df['VALOR_PREVISTO_ACTUALIZADO'] = df['VALOR_PREVISTO_ACTUALIZADO'].astype(float)
df.dtypes
```

```python
# Reemplazamos las comas por puntos.
df['VALOR_LANZADO'] = df['VALOR_LANZADO'].str.replace(',', '.')

# Convertir la columna al tipo float
df['VALOR_LANZADO'] = df['VALOR_LANZADO'].astype(float)
df.dtypes
```

```python
# Reemplazamos las comas por puntos.
df['PORCENTAJE_REALIZADO'] = df['PORCENTAJE_REALIZADO'].str.replace(',', '.')

# Convertir la columna al tipo float
df['PORCENTAJE_REALIZADO'] = df['PORCENTAJE_REALIZADO'].astype(float)
df.dtypes
```

```python
# Suponiendo que 'nombre_columna' es la columna que deseas convertir a fecha
df['FECHA_LANZAMIENTO'] = pd.to_datetime(df['FECHA_LANZAMIENTO'], format='%d/%m/%Y', errors='coerce')
df['FECHA_LANZAMIENTO'].unique()
```

```python
# Comprobamos valores nulos
df['AÑO_EJERCICIO'] = df['AÑO_EJERCICIO'].fillna(0)  # o usa el valor que prefieras para reemplazar nulos

# Convertir la columna de float a integer
df['AÑO_EJERCICIO'] = df['AÑO_EJERCICIO'].astype(int)
df.dtypes
```
